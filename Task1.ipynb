```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your data is in a CSV file named 'population_data.csv'
# and has columns like 'Country' and '2023 Population'
try:
    df = pd.read_csv('world population.csv')
except FileNotFoundError:
    print("Error: 'world population.csv' not found. Please make sure the file exists in the same directory.")
    exit()

# Sort by population in descending order and select the top 10
top_10_countries = df.sort_values(by='2023 population', ascending=False).head(10)

# Create the bar plot
plt.figure(figsize=(8, 4))  # Adjust figure size for better readability
plt.bar(top_10_countries['country'], top_10_countries['2023 population'], color='navy')
plt.xlabel('Country', fontsize=12)
plt.ylabel('population (2023)',fontsize=12)
plt.title('Top 10 Most Populated Countries (2023)',fontsize=12)
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("world population.csv")
# Check the first few rows
print(df.head())

# OPTIONAL: Clean or filter the data if necessary
# For example, let's filter for population in a specific year, like 2020:
df_2020 = df["2020 population"] 
# Plot a histogram of population in 2020
plt.figure(figsize=(12, 6))
sns.histplot(df_2020, bins=30, kde=True)
plt.title("World Population Distribution in 2020")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("world population.csv")

# Plot settings
plt.figure(figsize=(12, 6))

# Loop over all year columns (excluding 'Country' or other non-numeric)
for col in df.columns:
    if "population" in col.lower():
        sns.kdeplot(df[col].dropna(), label=col, fill=True, alpha=0.4)

plt.title("World Population Distribution Across Years")
plt.xlabel("Population")
plt.ylabel("Density")
plt.legend(title="Year")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
